```python
num_features = ['child_mort','exports','health','imports','income','inflation','life_expec','total_fer','gdpp']
data = data.iloc[1:,:]
for i in num_features:
    data[i] = data[i].astype(float)
print(data.info())
```

```python
#plotting the correlation matrix
plt.figure(figsize=(20,10))
sns.heatmap(data.corr(),annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
scaler = StandardScaler()
data[num_features] = scaler.fit_transform(data[num_features])
#data[num_features] = normalize(data[num_features])
print(data.head())
country = data['country']
```

```python
from sklearn.decomposition import PCA
pca = PCA()
data_pca = pd.DataFrame(pca.fit_transform(data[num_features]))

explained_variance = pca.explained_variance_ratio_
print("Explained Variance:")
print(explained_variance)
cumulative_variance = np.cumsum(np.round(explained_variance,decimals=4)*100)
print("Cumulative Explained Variance:")
print(cumulative_variance)

plt.figure(figsize=(10,5))
plt.plot(cumulative_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance (%)")
plt.title("Explained Variance")
plt.show()

plt.figure(figsize=(10,5))
plt.plot(explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance (%)")
plt.title("Explained Variance")
plt.show()
```

```python
data_pca = pd.DataFrame(pca.components_,columns=num_features)
plt.figure(figsize=(10,5))
sns.heatmap(data_pca,cmap='plasma',annot=True)
plt.title("Correlation Matrix: PCA Components vs Original Features")
plt.show()
```

```python
#Optimal Componenents
pca_opt = PCA(n_components=6)
data_pca_opt = pd.DataFrame(pca_opt.fit_transform(data[num_features]))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_pca_opt[0],data_pca_opt[1])
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA 1 vs PCA 2")
plt.show()

sns.pairplot(data_pca_opt)
plt.show()
```

```python
from sklearn.cluster import KMeans
#elbow method
wcss = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data_pca_opt)
    wcss.append(kmeans.inertia_)

#silhouette score method
from sklearn.metrics import silhouette_score
silhouette_scores = []
for i in range(2,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data_pca_opt)
    silhouette_scores.append(silhouette_score(data_pca_opt,kmeans.labels_))

# Elbow method
plt.figure(figsize=(10, 5))
plt.scatter(range(1, 10), wcss, marker='o', s=100, c='red')
plt.plot(range(1, 10), wcss, linestyle='dashed', markerfacecolor='blue')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()

# Silhouette score method
plt.figure(figsize=(10, 5))
plt.scatter(range(2, 10), silhouette_scores, marker='o', s=100, c='red')
plt.plot(range(2, 10), silhouette_scores, linestyle='dashed', markerfacecolor='blue')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,random_state=0)
res = kmeans.fit_predict(data_pca_opt)
kmeans_data = pd.concat([country,pd.Series(res,name='Cluster')],axis=1)
print(kmeans_data)

#list all the countries in cluster 0
print(kmeans_data[kmeans_data['Cluster']==0]['country'].values)

#list all the countries in cluster 1
print(kmeans_data[kmeans_data['Cluster']==1]['country'].values)

#list all the countries in cluster 2
print(kmeans_data[kmeans_data['Cluster']==2]['country'].values)

kmeans_data = pd.concat([data[num_features],kmeans_data],axis=1)
```

```python
# Scatter plot with distinct cluster colors
plt.figure(figsize=(10, 5))
scatter = plt.scatter(data_pca_opt[0], data_pca_opt[1], c=res, cmap='viridis', alpha=0.8)

plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("KMeans Clusters")
plt.colorbar(scatter, label='Cluster')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Assuming 'num_features' is a list of numeric features
for i in num_features:
    plt.figure(figsize=(12, 6))

    # Use a color palette for better aesthetics
    colors = sns.color_palette('viridis', n_colors=len(kmeans_data['Cluster'].unique()))

    sns.boxplot(x='Cluster', y=i, data=kmeans_data, palette=colors)
    plt.title(f'Boxplot of {i} vs Cluster')
    plt.xlabel('Cluster')
    plt.ylabel(i)
    plt.show()
```

```python
cluster_means = kmeans_data.groupby('Cluster')[num_features].mean()

plt.figure(figsize=(12, 6))
sns.heatmap(cluster_means, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Mean of Numeric Features Across Clusters')
plt.show()
```
